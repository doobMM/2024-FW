## üìå Q3. groupbyÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 2020ÎÖÑÏùò Ïò®ÎùºÏù∏ Ìï¥Ïô∏ÌåêÎß§ ÏÉÅÌíàÍµ∞ÏùÑ ÏßÄÏó≠Î≥Ñ Ìï©Í≥ÑÎ•º Íµ¨Ìï¥ Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî.

* 2020ÎÖÑÏùÄ ÏΩîÎ°úÎÇòÎ°ú Ïù∏Ìï¥ Ï†ÑÍµ≠ÎØºÏù¥ Ïñ¥Î†§ÏõÄÏùÑ Í≤™ÏóàÎçò ÏãúÍ∏∞ÏûÖÎãàÎã§. Îòê, ÏïÑÎßàÏ°¥ Îì±Ïùò Ìï¥Ïô∏ Ïò®ÎùºÏù∏ ÌåêÎß§Í∞Ä ÎäòÏñ¥ÎÇú ÏãúÍ∏∞Ïù¥Í∏∞ÎèÑ Ìï©ÎãàÎã§. Ïù¥ Îïå, Ìï¥Ïô∏ Ïñ¥ÎäêÏßÄÏó≠Ïóê Ïò®ÎùºÏù∏ ÌåêÎß§Í∞Ä ÏñºÎßàÎÇò ÎêòÏóàÎäîÏßÄÎ•º groupby() ÏôÄ unstack()ÏùÑ ÌôúÏö©Ìï¥ÏÑú Ìï©Í≥ÑÎ•º Íµ¨ÌïòÍ≥† Î∂ÑÏÑùÌï¥ Ï£ºÏÑ∏Ïöî! Îòê Î∂ÑÏÑùÌïú Í≤∞Í≥ºÎ•º Î≥¥Í≥† Ïñ¥Îñ§ Ïù∏ÏÇ¨Ïù¥Ìä∏Î•º ÏñªÏóàÎäîÏßÄÎèÑ Ìï®Íªò ÏûëÏÑ±Ìï¥ Ï£ºÏÑ∏Ïöî!
* Í∑∏ÎûòÌîÑÏùò ÏÉâÏÉÅ, Ïä§ÌÉÄÏùºÏùÄ ÏûêÏú†Î°≠Í≤å ÏõêÌïòÎäî Ïä§ÌÉÄÏùºÎ°ú ÏßÄÏ†ïÌï¥ÎèÑ Îê©ÎãàÎã§.

```python
# 2020ÎÖÑ Îç∞Ïù¥ÌÑ∞Îßå Í∞ÄÏ†∏ÏôÄÏ§çÎãàÎã§.
df_2020 = df[df["Ïó∞ÎèÑ"] == 2020].copy()
df_2020
```

```python
# Í¥ÄÏã¨ ÏóÜÎäî Îç∞Ïù¥ÌÑ∞Î•º Ï≥êÎÇ¥Í∏∞ Ï†Ñ, Ìïú Î≤à Îçî ÌôïÏù∏Ìï¥Ï§çÎãàÎã§.
print(df_2020["ÌåêÎß§Ïú†ÌòïÎ≥Ñ"].unique())
print("\n")

# 2020ÎÖÑÏóê ÏßëÏ§ëÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Î∂ÑÍ∏∞Î≥Ñ Î≥ÄÌôîÏ∂îÏÑ∏ Ïó≠Ïãú Ï†úÍ±∞Ìï©ÎãàÎã§
# Before
print(df_2020.info(verbose = False)) 
print("\n")

# After
df_2020.drop(["ÌåêÎß§Ïú†ÌòïÎ≥Ñ", "Ïó∞ÎèÑ", "Î∂ÑÍ∏∞", "ÏãúÏ†ê"], axis=1, inplace=True)
print(df_2020.info(verbose = False))
# Î©îÎ™®Î¶¨ Í∞êÏÜå Ï∂îÏ†Å : 31.5+ KB ‚Üí 15.8+ KB
```

```python
# groupbyÏôÄ unstackÏùÑ ÎèôÏãúÏóê ÏÇ¨Ïö©Ìï¥Ï£ºÍ∏∞ ÏúÑÌï¥ 
# 2Í∞úÏùò Ïª¨Îüº("Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ", "ÏÉÅÌíàÍµ∞Î≥Ñ")ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú groupby ÏàòÌñâ ÌõÑ ÌéºÏ≥êÏ§çÎãàÎã§.
df_Q3_raw = df_2020.groupby(["Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ", "ÏÉÅÌíàÍµ∞Î≥Ñ"])["Î∞±Îßå"].sum()
df_Q3_raw.unstack()
```

```python
# heatmapÏùÑ ÌÜµÌï¥ Î≥¥Í∏∞ ÏúÑÌï¥ ÏÉàÎ°úÏö¥ Î≥ÄÏàòÏóê Îã¥ÏïÑÏ§çÎãàÎã§.
df_Q3_corr = df_Q3_raw.unstack()

plt.figure(figsize=(15, 6))
plt.xticks(rotation=30) # ÏõêÏù∏ÎØ∏ÏÉÅ don't work
sns.heatmap(df_Q3_corr, cmap="coolwarm", annot=True, fmt=".0f", center = 5000, vmax = 30000)
```

* ‚òÖ Í∑πÍ∞íÏùò Ìé∏Ìñ•ÏÑ±Ïù¥ Ïª§ÏÑú ÎÇòÎ®∏ÏßÄ Ìï≠Î™©Îì§ Í∞ÑÏùò Ï∞®Ïù¥Í∞Ä Î≥ÄÎ≥ÑÎêòÏßÄ ÏïäÏäµÎãàÎã§.
* Îî∞ÎùºÏÑú Îã®ÏùºÌïú ÏÉâÍπî(cmap = "Reds")ÏùÑ ÌÜµÌï¥ Îß§ÌïëÌïòÎ©¥ÏÑú vmaxÎ•º ÌÜµÌï¥ ÏµúÎåìÍ∞íÏùÑ Ï†úÏñ¥ÌïòÍ±∞ÎÇò
* WK3 ÎØ∏ÏÖò ÎïåÏ≤òÎüº Ï†ÅÏ†àÌïú ÏÑºÌÑ∞Í∞íÍ≥º Ìï®Íªò coolwarmÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.

## üìå Q4. Ï£ºÏöî ÌåêÎß§ Íµ≠Í∞ÄÏôÄ ÏÉÅÌíàÍµ∞Ïóê ÎåÄÌï¥ 2021ÎÖÑÍπåÏßÄÏùò Ïó∞ÎèÑÎ≥Ñ Ïò®ÎùºÏù∏ ÏßÅÏ†ë ÌåêÎß§Ïï° Ìï©Í≥ÑÎ•º ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî.

* Ïò®ÎùºÏù∏ ÏßÅÏ†ë ÌåêÎß§Ïï°Ïù¥ ÎÜíÏùÄ ÏßÄÏó≠ÏùÄ "ÎØ∏Íµ≠", "Ï§ëÍµ≠", "ÏùºÎ≥∏" Ïù¥Î©∞, Ï£ºÏöî ÌåêÎß§ ÏÉÅÌíàÍµ∞ÏùÄ "ÏùòÎ•ò Î∞è Ìå®ÏÖò Í¥ÄÎ†®ÏÉÅÌíà", "ÌôîÏû•Ìíà", "ÏùåÎ∞ò¬∑ÎπÑÎîîÏò§¬∑ÏïÖÍ∏∞" ÏûÖÎãàÎã§. 2021ÎÖÑÍπåÏßÄÏùò Í∞Å ÏßÄÏó≠Î≥Ñ Ï†úÌíàÎ≥Ñ ÌåêÎß§Ïï°ÏùÑ ÏãúÍ∞ÅÌôî Ìï¥Ï£ºÏÑ∏Ïöî!
* ÏãúÍ∞ÅÌôî Í≤∞Í≥ºÏùò ÏÉâÏÉÅÏù¥ÎÇò Ïä§ÌÉÄÏùºÏù¥ÎÇò Ïä§ÌÉÄÏùºÏùÄ Îã¨ÎùºÎèÑ Í¥úÏ∞ÆÏäµÎãàÎã§.
 
üîΩ Ï∂úÎ†• ÏòàÏãú


```python
# Í¥ÄÏã¨Ïù¥ ÏóÜÎäî 'ÌåêÎß§Ïú†ÌòïÎ≥Ñ', 'ÏãúÏ†ê', 'Î∂ÑÍ∏∞' Ïª¨ÎüºÏùÑ Ï†úÏô∏ÌïòÍ≥† ÏÉàÎ°ú Í∞ÄÏ†∏ÏòµÎãàÎã§.
df_Q4 = df[["Íµ≠Í∞Ä(ÎåÄÎ•ô)Î≥Ñ", "ÏÉÅÌíàÍµ∞Î≥Ñ", "Î∞±Îßå", "Ïó∞ÎèÑ"]].copy()

# 'ÏÉÅÌíàÍµ∞Î≥Ñ' Ïª¨ÎüºÏóê ÎåÄÌï¥ÏÑúÎèÑ Í¥ÄÏã¨ÏûàÎäî Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂úÌï©ÎãàÎã§.
df_Q4 = df_Q4[(df_Q4["ÏÉÅÌíàÍµ∞Î≥Ñ"] == 'ÏùåÎ∞ò¬∑ÎπÑÎîîÏò§¬∑ÏïÖÍ∏∞') | 
             (df_Q4["ÏÉÅÌíàÍµ∞Î≥Ñ"] == 'ÏùòÎ•ò Î∞è Ìå®ÏÖò Í¥ÄÎ†®ÏÉÅÌíà') |
             (df_Q4["ÏÉÅÌíàÍµ∞Î≥Ñ"] == 'ÌôîÏû•Ìíà')]

# 'Ïó∞ÎèÑ' Ïª¨ÎüºÏóê ÎåÄÌï¥ÏÑúÎèÑ Í¥ÄÏã¨ÏûàÎäî Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂úÌï©ÎãàÎã§.
df_Q4 = df_Q4[~(df_Q4["Ïó∞ÎèÑ"] == 2022)]

# Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞Î•º ÎåÄÏ∂© Î¥ÖÎãàÎã§.
df_Q4.shape
```
