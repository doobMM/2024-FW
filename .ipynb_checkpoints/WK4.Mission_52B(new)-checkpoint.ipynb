mean) 2.5보다 작은 것에서 데이터가 2017.1/4에서 시작해 어느 해의 4분기까지 이어지지 않음을 알 수 있다.

## 📌 Q3. groupby를 사용하여 2020년의 온라인 해외판매 상품군을 지역별 합계를 구해 분석해 주세요.

* 2020년은 코로나로 인해 전국민이 어려움을 겪었던 시기입니다. 또, 아마존 등의 해외 온라인 판매가 늘어난 시기이기도 합니다. 이 때, 해외 어느지역에 온라인 판매가 얼마나 되었는지를 groupby() 와 unstack()을 활용해서 합계를 구하고 분석해 주세요! 또 분석한 결과를 보고 어떤 인사이트를 얻었는지도 함께 작성해 주세요!
* 그래프의 색상, 스타일은 자유롭게 원하는 스타일로 지정해도 됩니다.

```python
# 2020년 데이터만 가져와줍니다.
df_2020 = df[df["연도"] == 2020].copy()
df_2020
```

```python
# 관심 없는 데이터를 쳐내기 전, 한 번 더 확인해줍니다.
print(df_2020["판매유형별"].unique())
print("\n")

# 2020년에 집중하기 때문에 분기별 변화추세 역시 제거합니다
# Before
print(df_2020.info(verbose = False)) 
print("\n")

# After
df_2020.drop(["판매유형별", "연도", "분기", "시점"], axis=1, inplace=True)
print(df_2020.info(verbose = False))
# 메모리 감소 추적 : 31.5+ KB → 15.8+ KB
```

```python
# groupby와 unstack을 동시에 사용해주기 위해 
# 2개의 컬럼("국가(대륙)별", "상품군별")을 기준으로 groupby 수행 후 펼쳐줍니다.
df_Q3_raw = df_2020.groupby(["국가(대륙)별", "상품군별"])["백만"].sum()
df_Q3_raw.unstack()
```

```python
# heatmap을 통해 보기 위해 새로운 변수에 담아줍니다.
df_Q3_corr = df_Q3_raw.unstack()

plt.figure(figsize=(15, 6))
plt.xticks(rotation=30) # 원인미상 don't work
sns.heatmap(df_Q3_corr, cmap="coolwarm", annot=True, fmt=".0f", center = 5000, vmax = 30000)
```

* ★ 극값의 편향성이 커서 나머지 항목들 간의 차이가 변별되지 않습니다.
* 따라서 단일한 색깔(cmap = "Reds")을 통해 매핑하면서 vmax를 통해 최댓값을 제어하거나
* WK3 미션 때처럼 적절한 센터값과 함께 coolwarm을 사용합니다.

## 📌 Q4. 주요 판매 국가와 상품군에 대해 2021년까지의 연도별 온라인 직접 판매액 합계를 시각화 해주세요.

* 온라인 직접 판매액이 높은 지역은 "미국", "중국", "일본" 이며, 주요 판매 상품군은 "의류 및 패션 관련상품", "화장품", "음반·비디오·악기" 입니다. 2021년까지의 각 지역별 제품별 판매액을 시각화 해주세요!
* 시각화 결과의 색상이나 스타일이나 스타일은 달라도 괜찮습니다.
 
🔽 출력 예시


```python
# 관심이 없는 '판매유형별', '시점', '분기' 컬럼을 제외하고 새로 가져옵니다.
df_Q4 = df[["국가(대륙)별", "상품군별", "백만", "연도"]].copy()

# '상품군별' 컬럼에 대해서도 관심있는 데이터만 추출합니다.
df_Q4 = df_Q4[(df_Q4["상품군별"] == '음반·비디오·악기') | 
             (df_Q4["상품군별"] == '의류 및 패션 관련상품') |
             (df_Q4["상품군별"] == '화장품')]

# '연도' 컬럼에 대해서도 관심있는 데이터만 추출합니다.
df_Q4 = df_Q4[~(df_Q4["연도"] == 2022)]

# 데이터 크기를 대충 봅니다.
df_Q4.shape
```

```python
# subplot의 타이틀 출력 밑밥
print(df_Q4["국가(대륙)별"].unique())
print(df_Q4["국가(대륙)별"].unique()[0])
```

```python
# '국가(대륙)별' 컬럼에 대해서 관심있는 데이터만 가져옵니다.
df_Q4_US = df_Q4[df_Q4["국가(대륙)별"] == '미국'].copy()
df_Q4_CN = df_Q4[df_Q4["국가(대륙)별"] == '중국'].copy()
df_Q4_JP = df_Q4[df_Q4["국가(대륙)별"] == '일본'].copy()

# ★
fig, ax = plt.subplots(1, 3, sharey = True, figsize=(15, 4))

df_Q4_US_plot = sns.lineplot(data=df_Q4_US, x="연도", y="백만", hue="상품군별", estimator = 'sum', errorbar=None, marker='o', 
                             ax = ax[0], linewidth = 2, legend = False)
df_Q4_CN_plot = sns.lineplot(data=df_Q4_CN, x="연도", y="백만", hue="상품군별", estimator = 'sum', errorbar=None, marker='o', 
                             ax = ax[1], linewidth = 2, legend = False)
df_Q4_JP_plot = sns.lineplot(data=df_Q4_JP, x="연도", y="백만", hue="상품군별", estimator = 'sum', errorbar=None, marker='o', 
                             ax = ax[2], linewidth = 2)

# 각 subplot의 오른쪽과 위쪽 선은 안 보이게 합니다. 
for i in range(3):
    sns.despine(ax=ax[i], top=True, right=True)
    ax[i].title.set_text('국가(대륙)별 = ' + df_Q4["국가(대륙)별"].unique()[i])


plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left', borderaxespad=0, fontsize = 9, title = '상품군별', frameon = False)


plt.show() 
```
